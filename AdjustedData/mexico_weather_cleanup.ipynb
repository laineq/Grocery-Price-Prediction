```python
import pandas as pd

df = pd.read_csv("/Users/jli624/Desktop/cmpt733/project/mexico_weather.csv")

df.head()
```

```python
# rename first column
df = df.rename(columns={"PERIOD": "date"})

# change date format yyyy-mm-dd -> yyyy-mm
df["date"] = df["date"].str[:7]

# keep only selected states
df = df[df["STATE_ID"].isin([16, 14, 15, 25])]

# drop state_id column
df = df.drop(columns=["STATE_ID"])

# export csv
df.to_csv(
    "/Users/jli624/Desktop/cmpt733/project/mexico_weather_adjusted.csv",
    index=False
)
```
