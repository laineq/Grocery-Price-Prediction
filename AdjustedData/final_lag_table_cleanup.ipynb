```python
import pandas as pd

avocado = pd.read_csv("/Users/jli624/Desktop/cmpt733/project/avocado_final_selective_log.csv")
tomato = pd.read_csv("/Users/jli624/Desktop/cmpt733/project/tomato_final_selective_log.csv")

tomato.head()
```

```python
# rename columns
avocado = avocado.rename(columns={
    avocado.columns[1]: "Mean Temperature",
    avocado.columns[2]: "Precipitation",
    avocado.columns[3]: "MXNCAD Exchange Rate",
    avocado.columns[4]: "USDCAD Exchange Rate",
    avocado.columns[5]: "Integrated Gas Price",
    avocado.columns[6]: "Import Quantity",
    avocado.columns[7]: "Avocado Price"
})

# rounding
avocado["Mean Temperature"] = avocado["Mean Temperature"].round(1)
avocado["Precipitation"] = avocado["Precipitation"].round(1)
avocado["MXNCAD Exchange Rate"] = avocado["MXNCAD Exchange Rate"].round(5)
avocado["USDCAD Exchange Rate"] = avocado["USDCAD Exchange Rate"].round(4)
avocado["Integrated Gas Price"] = avocado["Integrated Gas Price"].round(5)
avocado["Avocado Price"] = avocado["Avocado Price"].round(2)

avocado.to_csv(
    "/Users/jli624/Desktop/cmpt733/project/avocado_final_lag_table.csv",
    index=False
)
```
