```python
import pandas as pd

imported = pd.read_csv("/Users/jli624/Desktop/cmpt733/project/avocado_imported_from_mexico.csv")

imported.head()
```

```python
# create date column in yyyy-mm format
imported["date"] = (
    imported["refYear"].astype(str) + "-" +
    imported["refMonth"].astype(str).str.zfill(2)
)

# keep only date and qty
imported_clean = imported[["date", "qty"]]

# convert qty to integer
imported_clean["qty"] = imported_clean["qty"].astype(int)

# export CSV
imported_clean.to_csv(
    "/Users/jli624/Desktop/cmpt733/project/avocado_import.csv",
    index=False
)
```

```python
tomato = pd.read_csv("/Users/jli624/Desktop/cmpt733/project/tomato_import_mexico_us.csv")
tomato.head()
```

```python
# create date column yyyy-mm
tomato["date"] = (
    tomato["refYear"].astype(str) + "-" +
    tomato["refMonth"].astype(str).str.zfill(2)
)

# keep relevant columns
tomato_clean = tomato[["date", "partnerDesc", "qty"]]

# rename column
tomato_clean = tomato_clean.rename(columns={"partnerDesc": "partner"})

# convert qty to integer
tomato_clean["qty"] = tomato_clean["qty"].astype(int)

# export CSV
tomato_clean.to_csv(
    "/Users/jli624/Desktop/cmpt733/project/tomato_import.csv",
    index=False
)
```
